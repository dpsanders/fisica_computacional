## Descripción de los proyectos finales 

El proyecto final constituye una parte suma importante del curso. Provee la oportunidad para que tú muestres que hayas captado la esencia del curso, que sepas programar y que sepas investigar por tu cuenta, y para que te adentres en un tema que te guste.

Consiste en desarrollar un proyecto en pareja (máximo tres personas en circunstancias especiales), sobre un tema que **no** hayamos tratado en el curso, pero que es del mismo nivel, o es una extensión de lo que hemos visto, y que ocupa el mismo tipo de herramientas computacionales y métodos numéricos. Para buscar un tema adecuado, te recomendamos buscar en los libros del curso, por ejemplo el de Newman, de Giordano, etc.

## Estructura del proyecto

Hay que entregar un notebook con código en Julia que presente brevemente el **problema físico** que resolver, incluyendo las **ecuaciones** correspondientes, tu **código** (resaltando los puntos importantes, delicados y/o interesantes; el código se puede incluir como un script `.jl` aparte si prefieres), unas **figuras y animaciones**, y **conclusiones físicas**. También debes incluir las **referencias** de los libros (y/o artículos etc.) que hayas utilizado.

También habrá una **presentación oral de 15 minutos**, en la cual deben explicar **brevemente** el contenido del proyecto, y en la cual dos personas participantes deben mostrar que hayan contribuido de formas iguales al desarrollo del proyecto. Estas presentaciones orales se llevarán a cabo en la segunda semana de exámenes. Es necesario que **practiquen la presentación de antemano** para asegurarse de que **dure (sólo) 15 minutos**.

Manda los nombres de las dos personas en el equipo, así como una sugerencia de tema de proyecto a los ayudantes **antes del 1 de noviembre**. De ser aceptado el tema, se colocará en una lista en este notebook. No se permite repetir temas, así que ¡vale la pena escoger primero!
